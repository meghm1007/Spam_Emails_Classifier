```python
#Import Libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
df = pd.read_csv('Copy of mail_data.csv')
```

```python
df
```

```python
data = df.where((pd.notnull(df)),'')
```

```python
data.head()
```

```python
data.shape
```

```python
data.loc[data['Category'] == 'spam', 'Category',] = 0 #convert spams to 0
data.loc[data['Category'] == 'ham', 'Category',] = 1 #convert not spams (ham) to 1
```

```python
X = data['Message']
y = data['Category']
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=3)
print(df.shape)
print(data.shape)
print('---------')
print(X.shape)
print(X_train.shape)
print(X_test.shape)
print('---------')
print(y.shape)
print(y_test.shape)
print(y_train.shape)
```

```python
feature_extraction = TfidfVectorizer(min_df = 1, stop_words='english')
X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)

y_train = y_train.astype('int')
y_test = y_test.astype('int')
```

```python
print(X_train)
```

```python
print(X_train_features)
```

```python
model = LogisticRegression()
model.fit(X_train_features, y_train)
```

```python
pred = model.predict(X_train_features)
accuracy = accuracy_score(y_train, pred)
```

```python
accuracy
```

```python
def input_mail(mailtext)->str:
  input_data_features = feature_extraction.transform(mailtext)
  final_prediction = model.predict(input_data_features)
  if final_prediction[0]==1:
    return 'Not Spam'
  return 'Spam'
```
